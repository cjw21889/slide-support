## Miscellaneous

### Args and Kwargs

```python
def my_func(required, *args, **kwargs):
    print('required: ', required)
    if args:
        print('my args: ', args)
    if kwargs:
        print('my kwargs:', kwargs)
```

```python
# kwargs keys in on = to seperate them from args
# and packs them into a dictionary
my_func(1, 2, 3, a=4, b=5)
```

```python
# We can create a dictionary with all our arguments  
my_dict = {'a':4, 'b':5}
# if we pass it directly, it will be an arg
my_func(1, 2, 3, my_dict)
```

```python
# Or un-pack the dict before it goes in
# and the function will re-pack them
my_func(1, 2, 3, **my_dict)
```

```python
# same for a list 
my_list = [2, 3]
my_func(1, my_list, **my_dict)
```

```python
# unpack them, function repacks them
my_func(1, *my_list, **my_dict)
```

### Wrappers

We are basically just passing our function into another function which updates it.  You will use some, but YOU WILL NEVER HAVE TO DEFINE YOUR OWN IF YOU DON"T WANT TO

```python
# Manually applying the decorator

# define decorator
def my_decorator(func):
    def wrapper(*args):
        print('I added this to happen before your function')
        func(*args)
        print("I added this to happen after your function")
    return wrapper

# define function
def my_round(flt):
    print(round(flt))

# re-assign function to be decorated version of function
my_round = my_decorator(my_round)

```

```python
# call the re-assigned function
my_round(3.14)
```

```python
# using the @keyword
# again define decorator
def my_decorator_2(func):
    def wrapper(*args):
        print('I added this to happen before your function')
        func(*args)
        print("I added this to happen after your function")
    return wrapper

# pre-empt function definition with @<decorator name>
@my_decorator_2
def my_round_2(flt):
    print(round(flt))
    return round(flt)
```

```python
# Call the function
my_round_2(3.14)
```
