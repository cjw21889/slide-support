## DataFrame

```python
# Can create dataframes directly with values
df = pd.DataFrame(
    [[4, 7, 10],
     [5, 8, 11],
     [6, 9, 12]],
    index=['row_1', 'row_2', 'row_3'],
    columns=["col_a", "col_b", "col_c"]
)
df
```

```python
# Or by combining multiple series
apples  = pd.Series(data=[1, 2, 3], index=["id1", "id2", "id3"])
oranges = pd.Series(data=[4, 5, 6], index=["id1", "id2", "id3"])
d = {
    "apples": apples,
    "oranges": oranges,
}
pd.DataFrame(d)
```

## Exploratory Analysis

```python
# Loading in our csv file
countries_df = pd.read_csv('countries.csv', decimal=',')
```

```python
countries_df.shape # => Tuple representing the dimensionality of the DataFrame
```

```python
countries_df.info() # => gives a description of each column for how many NON null values
```

```python
countries_df.isnull().sum() # => can use isnull().sum() to get total ACTUAL null values
```

```python
countries_df.describe() # => describe returns stats for all numerical columns
```

```python
# head / tail will let you preview top and bottom of df
countries_df.head(2)
# countries_df.tail(2)
```

```python
# single bracket column indexing returns a series
countries_df["Country"]
#  double bracket retrns a dataframe
countries_df[["Country"]]

# type(countries_df[["Country"]])
```

```python
# Can index multiple columns using double brackets
countries_df[["Country", "Region"]]
```

```python
# .loc to index rows / columns
countries_df.loc[0:5, ["Country", "Region"]] # From row index 0 to 5 (included)
```

## Boolean Indexing

```python
# The old slow way to index
big_countries = []
for index, country in countries_df.iterrows():
    if country['Population'] >= 1_000_000_000:
        big_countries.append(country)
pd.DataFrame(big_countries)
```

```python
# the beautiful quick way to index
countries_df[countries_df["Population"] >= 1_000_000_000]
```

```python
# as your boolean mask equations get longer you can assign them to variables
american = countries_df['Region'].str.contains('AMER')
# and then pass the variable as in index
countries_df[american]
```

```python
# isin method must match exactly
countries_df[countries_df["Region"].isin(["WESTERN EUROPE", "EASTERN EUROPE"])]
```

```python
# here we see the region needs to be stripped 
countries_df['Region'].unique()
```

```python
# we use the str.strip() method and assign it back to the region column
countries_df["Region"] = countries_df["Region"].str.strip()
```

```python
# now everything is clean
countries_df['Region'].unique()
```

```python
# and we can index
countries_df[countries_df["Region"].isin(["WESTERN EUROPE", "EASTERN EUROPE"])]
```

```python
# or use ~ to get inverse results
countries_df[~countries_df["Region"].isin(["WESTERN EUROPE", "EASTERN EUROPE"])]
```

## Re-indexing

```python
# we can also pass the method as an arugment to the map() function to apply to to the column
countries_df['Country'] = countries_df['Country'].map(str.strip)
# set index will remove our default numerical index and replace with our column values
countries_df.set_index('Country', inplace=True)
```

```python
countries_df.head()
```

```python
# Now we can index rows by the country name
countries_df.loc['France':'Germany', ['Region', 'Population']]
```

## Sorting

```python
# We can also sort by the new indexs a-z or z-a
countries_df.sort_index(ascending=False)
```

```python
# or use sort values to sort by a specific column
countries_df.sort_values(by='Population', ascending=False)
```

```python
# We can also tell the sort function where to place rows with null values
countries_df.sort_values(by='GDP ($ per capita)', na_position='first')
```

## Grouping

```python
# our groupby method returns an object NOT an actual dataframe
regions = countries_df.groupby('Region')
regions
```

```python
# from that object we select our columns and apply the method we want
regions[['Population', 'Area (sq. mi.)']].sum()
```

```python
# since a dataframe is returned after aggregating we can then chain other methods to it
regions[['Population', 'Area (sq. mi.)']].sum().sort_values('Population', ascending=False)
```
