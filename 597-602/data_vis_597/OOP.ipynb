# OOP Subplots 

```python
import matplotlib.pyplot as plt
```

```python
# define x and y datasets
years_x = [1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015]
total_y = [1243, 1543, 1619, 1831, 1960, 2310, 2415, 2270, 1918]
coal_y = [823, 1136, 1367, 1547, 1660, 1927, 1983, 1827, 1352]
gas_y = [171, 200, 166, 175, 228, 280, 319, 399, 529]
```

Create a 2 column subplot with the right spine missing from the first plot and the bottom spine missing from the second plot

```python
# Start a figure
fig = plt.figure(figsize=(10,3))

# First subplot
ax1 = fig.add_subplot(1,2,1)
ax1.plot(years_x, coal_y, label="coal")
ax1.plot(years_x, gas_y, label = "gas")
ax1.set_title('coal vs. gas')
ax1.legend()

# Second subplot
ax2 = fig.add_subplot(1,2,2)
ax2.plot(years_x, total_y, c='black')
ax2.set_title('all energies')

ax1.spines['right'].set_color(None)
ax2.spines['bottom'].set_color(None)

# Global figure methods
fig.suptitle('US electricity CO2 emissions')
plt.show()
```

## De-constructing Subplots

```python
# create our figure
fig = plt.figure(figsize=(10,3))

# then use .add_subplot to put on our figure one at a time
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.spines['right'].set_color(None)
ax2.spines['bottom'].set_color(None)
```

```python
# Or we can use .subplots to create figure and subplots all at once
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,3))
ax1.spines['right'].set_color(None)
ax2.spines['bottom'].set_color(None)
```

```python
# Lets look at what returns in the second position
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10,5))
```

```python
# its a numpy array matching out shape
display(type(axes))
axes.shape
```

```python
# our first row contains 3 subplots objects
axes[0]
```

```python
# we can access the individual axes through numpy index notation
axes[0,0]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15,10))
axes[0,0].plot(years_x, gas_y)
axes[0,0].set_title('Gas Production View')
axes[0,1].plot(years_x, coal_y)
axes[0,1].set_title('Coal Production View')
axes[1,0].plot(years_x, total_y)
axes[1,0].set_title('Total Production View')
# Combined for our 4th plot
axes[1,1].plot(years_x, total_y, label="total")
axes[1,1].plot(years_x, gas_y, label="gas")
axes[1,1].plot(years_x, coal_y, label="coal")
axes[1,1].legend()
axes[1,1].set_title("Combined View");

axes[0,0].spines['right'].set_color(None)
axes[1,0].spines['right'].set_color(None)
axes[0,1].spines['bottom'].set_color(None)
axes[1,1].spines['bottom'].set_color(None)
```
