```python
import matplotlib
%matplotlib inline
import numpy as np
import pandas as pd
from music import fetch_lyrics 
```

```python
tracks_df = pd.read_csv('data/spotify_2017.csv')
```

```python
tracks_df.head()
```

```python
tracks_df['lyrics'] = ""
```

```python
%%time
for index, row in tracks_df.head(5).iterrows():
    print(f"Fetching lyrics for {row['artists']} - {row['name']}")
    lyrics = fetch_lyrics(row['artists'], row['name'])
    tracks_df.loc[index, 'lyrics'] = lyrics
```

```python
tracks_df.head()
```

## Apply

```python
tracks_df['lyrics'] = tracks_df.apply(lambda x: fetch_lyrics(x['artists'], x['name']), axis=1)
```
