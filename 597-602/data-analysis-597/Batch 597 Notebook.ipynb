## Vectorized Operations 

```python
my_list = [
    [6, 5],
    [1, 3],
    [5, 6],
    [1, 4],
    [3, 7],
    [5, 8],
    [3, 5],
    [8, 4],
]
```

```python
sums = []
for row in my_list:
    sums.append(row[0] + row[1]) # standard integer "+" operator
sums
```

```python
A = np.array(my_list)
my_sum = A[:,0] + A[:,1] # Vectorial "+" operator
my_sum
```

```python
np.sum(A, axis=0)  
# np.sum(A, axis=1)
```

```python
A = np.random.rand(10000, 1000)
a = A.tolist()
```

```python
%%time
total = 0
for row in a:
    for n in row:
        total += n
round(total, 2)
```

```python
%%time
round(np.sum(A), 2)
```

## Boolean Indexing

```python
big_countries = []
for index, country in countries_df.iterrows():
    if country['Population'] >= 1_000_000_000:
        big_countries.append(country)
pd.DataFrame(big_countries)
```

```python
countries_df["Population"] >= 1_000_000_000
```

```python
countries_df[countries_df["Population"] >= 1_000_000_000]
```

```python
countries_df['Region'].unique()
```

```python
american = countries_df['Region'].str.contains('AMER')
countries_df[american]
```

```python
big_eu = countries_df["Region"].isin(["WESTERN EUROPE", "EASTERN EUROPE"]) & (countries_df['Population'] > 10_000_000)
countries_df[big_eu]
```

```python
countries_df["Region"] = countries_df["Region"].str.strip()
```

```python
countries_df['Region'].unique()
```

```python
big_europe = (countries_df["Region"] == "WESTERN EUROPE") & (countries_df['Population'] > 10_000_000)
countries_df[big_europe]
```

## Re-indexing

```python
countries_df['Country'] = countries_df['Country'].str.strip() #(str.strip)
countries_df.set_index('Country', inplace=True)
```

```python
countries_df.loc['Algeria']
```
