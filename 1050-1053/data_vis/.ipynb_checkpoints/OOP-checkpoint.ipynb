# Object Oriented Subplotting

```python
import matplotlib.pyplot as plt
years_x = [1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015]
total_y = [1243, 1543, 1619, 1831, 1960, 2310, 2415, 2270, 1918]

coal_y = [823, 1136, 1367, 1547, 1660, 1927, 1983, 1827, 1352]
gas_y = [171, 200, 166, 175, 228, 280, 319, 399, 529]
```

```python
# Start a figure object
fig = plt.figure(figsize=(10,3))

# call methods on object
fig.suptitle('US electricity CO2 emissions')

# Assign first object subplot
ax1 = fig.add_subplot(1,2,1)
# assign 2 plots inside
ax1.plot(years_x, coal_y, label="coal")
ax1.plot(years_x, gas_y, label = "gas")
# assign a title and legened to the subplot
ax1.set_title('coal vs. gas')
ax1.legend()

# Assign second object subplot
ax2 = fig.add_subplot(1,2,2)
# assign a single plot and title
ax2.plot(years_x, total_y, c='black')
ax2.set_title('all energies');
```

How do we get access to each spine?

```python
# We can access directly from the individual created variables

display([i for i in ax1.spines.items()])


```

How do we hide the **right** spine from the **left** graph and the **bottom** spine from the **right** graph?

```python
# Start a figure object
fig = plt.figure(figsize=(10,3))

# call methods on object
fig.suptitle('US electricity CO2 emissions')

# Assign first object subplot
ax1 = fig.add_subplot(1,2,1)
# assign 2 plots inside
ax1.plot(years_x, coal_y, label="coal")
ax1.plot(years_x, gas_y, label = "gas")
# assign a title and legened to the subplot
ax1.set_title('coal vs. gas')
ax1.legend()

# Assign second object subplot
ax2 = fig.add_subplot(1,2,2)
# assign a single plot and title
ax2.plot(years_x, total_y, c='black')
ax2.set_title('all energies');

# Since the axes objects are assigned to variables, we can change them at any point in the code
ax1.spines['right'].set_color(None)
ax2.spines['bottom'].set_color(None)
```

## De-constructing subplots

What did we do above?

```python
# create our figure
fig = plt.figure(figsize=(10,3))

# then use .add_subplot to put on our figure one at a time
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

# and we can manipulate each individual axes from their variable name
ax1.spines['right'].set_color(None)
ax2.spines['bottom'].set_color(None)
```

Or we can use `.subplots()` to create figure and all subplots at once

```python
# Here saving the whole thing into a single variable "axes"
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10,5))
```

But what is this new axes variable?

```python
# its a numpy array matching our shape, 2 rows, 3 columns
display(type(axes))
display(axes.shape)
axes
```

```python
# first row has 3 columns
axes[0]
```

```python
# can index just like we would a normal numpy array
axes[0,0]
```

Let's see it in action

```python
# create figure with 2 rows of 2 subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15,10))

# Master title
fig.suptitle('CO2 emmissions')

# row 1 column 1 gets gas
axes[0,0].plot(years_x, gas_y)
axes[0,0].set_title('Gas Production View')

# row 1 column 2 gets coal
axes[0,1].plot(years_x, coal_y)
axes[0,1].set_title('Coal Production View')

# row 2 column 1 gets total
axes[1,0].plot(years_x, total_y)
axes[1,0].set_title('Total Production View')

# row 2 column 2 gets everything combined
axes[1,1].plot(years_x, total_y, label="total")
axes[1,1].plot(years_x, gas_y, label="gas")
axes[1,1].plot(years_x, coal_y, label="coal")
axes[1,1].legend()
axes[1,1].set_title("Combined View");

# Can continue to hide spines like we did before
axes[0,0].spines['right'].set_color(None)
axes[1,0].spines['right'].set_color(None)
axes[0,1].spines['bottom'].set_color(None)
axes[1,1].spines['bottom'].set_color(None)
```

# BONUS - Mapping functions to arrays reminder

```python
import math
import numpy as np
x = np.arange(1,11)

print('X: ', x, '\n')

# manually loop through all of the x array and call the function on each element
lc = [math.sqrt(i) for i in x]

print('LIST COMP: ', lc , '\n')

# OR Pass the function and the whole array into the map function
m = list(map(math.sqrt, x))
print('MAP: ', m)
```
