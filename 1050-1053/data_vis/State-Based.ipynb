# State Based Subplotting

```python
import matplotlib.pyplot as plt
years_x = [1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015]
total_y = [1243, 1543, 1619, 1831, 1960, 2310, 2415, 2270, 1918]

coal_y = [823, 1136, 1367, 1547, 1660, 1927, 1983, 1827, 1352]
gas_y = [171, 200, 166, 175, 228, 280, 319, 399, 529]
```

```python
# State a figure
plt.figure(figsize=(10,3))

# Global pyplot method
plt.suptitle('US electricity CO2 emissions')

# First stated subplot
plt.subplot(1,2,1)
# place 2 plots inside
plt.plot(years_x, coal_y, label="coal")
plt.plot(years_x, gas_y, label = "gas")
# add a title and legend to the subplot
plt.title('coal vs. gas')
plt.legend()

# Second stated subplot
plt.subplot(1,2,2) 
# place a single plot and title
plt.plot(years_x, total_y, label="total", c='black')
plt.title("all energies");
```

How do we get access to each spine?

```python
# Need to create an axes object to pull the spines from
plt.figure(figsize=(10,3))
ax = plt.gca()

[print(i) for i in ax.spines.items()]
plt.close()
```

How do we hide the **right** spine from the **left** graph and the **bottom** spine from the **right** graph?

```python
# State a figure
plt.figure(figsize=(10,3))

# Global pyplot method
plt.suptitle('US electricity CO2 emissions')

# First stated subplot
plt.subplot(1,2,1)
# place 2 plots inside
plt.plot(years_x, coal_y, label="coal")
plt.plot(years_x, gas_y, label = "gas")
# add a title and legend to the subplot
plt.title('coal vs. gas')
plt.legend()
# Grab axes before next subplot is created and set spine color
ax = plt.gca()
ax.spines['right'].set_color(None)

# Second stated subplot
plt.subplot(1,2,2) 
# place a single plot and title
plt.plot(years_x, total_y, label="total", c='black')
plt.title("all energies");
# Create 2nd axes object and set spine color 
ax1 = plt.gca()
ax1.spines['bottom'].set_color(None);

```
